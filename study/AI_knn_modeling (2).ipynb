```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier

heart = pd.read_csv("heart.csv", index_col=0)
heart.dropna(0,how='any',inplace=True)

heart["ChestPain"] = LabelEncoder().fit_transform(heart["ChestPain"])
heart["Thal"] = LabelEncoder().fit_transform(heart["Thal"])

print(heart.ChestPain)

features = [['Age', 'RestBP', 'Chol', 'Fbs', 'RestECG', 'MaxHR', 'ExAng', 'Oldpeak', 'Slope', 'Ca']]
for feature in features:
    heart[feature] = MinMaxScaler().fit_transform(heart[feature])
    
X = heart.iloc[:,:-1]
#y = heart.iloc[:,-1:]
y = np.array(heart.iloc[:,-1:]).ravel()

trainData, testData, trainLabel, testLabel = train_test_split(X, y, random_state=0)

for i in range(1,30,3):
    clf = KNeighborsClassifier(n_neighbors=i)
    clf.fit(trainData, trainLabel)
    print("Test set accuracy with k = {:} : {:.2f}".format(i, clf.score(testData, testLabel)))
```
