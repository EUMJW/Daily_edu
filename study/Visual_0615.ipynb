```python
# 임의로 2번, 3번 행의 모든 데이터가 누락 값이 되도록 변경
concrete.iloc[2:4] = np.nan
concrete[2:4]
```

```python
# 누락 값을 특정 값으로 채움
concrete.fillna(100.)
```

```python
# 누락 값을 평균 값으로 채움
concrete.fillna(concrete.mean(0))
```

```python
from numpy.random import randn
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(), 'g', label='one')
ax.plot(randn(1000).cumsum(), 'b--', label='two')
ax.plot(randn(1000).cumsum(), 'r.', label='three')
ax.legend(loc='best')
```
