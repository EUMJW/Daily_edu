```python
import csv

f = open('생태통로.csv', 'r')

read = csv.reader(f)

a=[]

for i,line in enumerate(read):
    if i != 0:
        x=float(line[6])+float(line[7])/60+float(line[8])/3600
        y=float(line[9])+float(line[10])/60+float(line[11])/3600
        a.append([x,y])

f.close()
```

```python
a
```

```python
import pandas as pd
dataframe = pd.DataFrame(a)
```

```python
dataframe
```

```python
dataframe.to_csv('생태.csv',header=False, index=False)
```
