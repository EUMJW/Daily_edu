```python
def solution(n):
    answer=''

    n=n-1

    while n >=0:
        if (n % 3) == 0:
            answer = '1'+answer
        elif (n % 3) == 1:
            answer = '2'+answer
        elif (n % 3) == 2:
            answer = '4'+answer
        n=n//3 -1

            
    return answer
```

```python
for i in range(1,20):
    print(solution(i))
```

```python
answer =''
def a(n):
    global answer
    if (n-1) %3 == 0:
        if answer != None:
            answer = answer+'1'
        else:
            answer = '1'
    elif (n-1) %3 == 1:
        if answer != None:
            answer = answer+'2'
        else:
            answer = '2'
    elif (n-1) %3 == 2:
        if answer != None:
            answer = answer+'4'
        else:
            answer = '4'
    elif (n) >3:
        n=((n)//3)
        a(n)
    return answer
```

```python
def pibo(n):
    if n == 1:
        return 1
    if n == 2:
        return 1
    if n>2:
        return (pibo(n-2) + pibo(n-1))
```

```python
import requests
from bs4 import BeautifulSoup



r = requests.get('https://en.wikipedia.org/wiki/Main_Page')

r

soup = BeautifulSoup(r.text,'lxml')
urls=[]

for url in soup.find_all("a"):
    if 'href' in url.attrs:
        if url.attrs['href'].startswith('/wiki/') == True:
            urls.append('https://en.wikipedia.org'+url.attrs['href'])

            
            
print(len(urls))           
print(urls)

# wiki_h=[]
# for i in wiki_h:
#     if 'wiki' in str(i):
#         wiki_h.append(wiki_h)
        
# len(wiki_h)

```

```python
import time

my_real_urls = []

for url in urls:
    r = requests.get(url)
    time.sleep(5)
    if r.status_code ==200:
        print("ok", end=' ')
        my_real_urls.append(url)
        
print(len(my_real_urls), my_real_urls)
```

```python
import requests
from bs4 import BeautifulSoup



r = requests.get('https://en.wikipedia.org/wiki/Main_Page')

r

soup = BeautifulSoup(r.text,'lxml')
portal_urls=[]

for url in soup.find_all("a"):
    if 'href' in url.attrs:
        if url.attrs['href'].startswith('/wiki/Portal') == True:
            portal_urls.append('https://en.wikipedia.org'+url.attrs['href'])

            
            
print(len(portal_urls))           
print(portal_urls)
```

```python
def urlsss(httpp):
    r = requests.get(httpp)
    
    soup = BeautifulSoup (r.text, 'lxml')
    
    for url in soup.find_all("a"):
        if 'href' in url.attrs:
            if url.attrs['href'].startswith('/wiki/Portal') == True:
                if 'https://en.wikipedia.org'+url.attrs['href'] not in portal_urls:
                    portal_urls.append('https://en.wikipedia.org'+url.attrs['href'])
```
